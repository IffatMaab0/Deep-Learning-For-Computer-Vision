```python
import tensorflow as tf
```

```python
x = [1,2,3]
tf.expand_dims(x, axis=0)
```

```python
x = tf.constant([-5, 3, -2])

tf.math.abs(x)
```

```python
x = tf.constant([2,3,4])

tf.math.square(x)
```

```python
x = tf.constant([4.0, 9.0, 16.0])

tf.math.sqrt(x)
```

```python
a = tf.constant([1,5,3])
b = tf.constant([2,4,10])

tf.math.maximum(a,b)
```

```python
x = tf.constant([1,2,3,4])

tf.reduce_sum(x)
```

```python
x = tf.constant([
    [1,2],
    [3,4]
])

tf.reduce_sum(x)
```

```python
x = tf.constant([5,20,8])
tf.math.argmax(x)
```
